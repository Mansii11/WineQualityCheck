```python
# Spltting dataset into training and testing dataset
```

```python
y=df['quality'].values
```

```python
X=df.drop(['quality'],axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,stratify=y)
```

```python
knn=KNeighborsClassifier(n_neighbors=33)
knn.fit(X_train,y_train)
knn.predict(X_test)
train_score=knn.score(X_train,y_train)
test_score=knn.score(X_test,y_test)
```

```python
print("KNN training accuracy is: ",train_score)
```

```python
print('KNN model accuracy score: ',test_score)
```

```python
# using cross validation method
```

```python
from sklearn.model_selection import cross_val_score
knn_algo=KNeighborsClassifier(n_neighbors=29)
cv_result=cross_val_score(knn_algo,X,y,cv=5)
cv_result
print('Cross validation model accuracy score: ',np.mean(cv_result))
```

```python
#using logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
logreg = LogisticRegression(multi_class='multinomial',solver ='newton-cg')
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print('Confusion matrix: \n\n',confusion_matrix(y_test,y_pred))

print('\nClassification report: \n\n',classification_report(y_test,y_pred,zero_division=1))
print('Accuracy using Logistic Regression is ',logreg.score(X_test,y_test))
```
